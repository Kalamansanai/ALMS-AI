```python
from lib.configreader import Config
from lib.trainer import Trainer
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np

# Train the model

cnf = Config()
trainer =  Trainer(cnf)
trainer.load_dataset()
trainer.start()

# Test for accuracy on the entire training data

image_data = trainer.train_data
label_data = trainer.train_labels

model = tf.keras.models.load_model("model.h5")
index = 0
counter = 0

for img in image_data:
 
    prediction = model.predict(np.array([img]))[0]
    prediction = [1 if x > 0.5 else 0 for x in prediction]
    prediction = np.array(prediction)
    
    if np.array_equal(prediction, label_data[index]):
        counter += 1
        
    index += 1

accuracy = (counter / image_data.shape[0]) * 100
print("Full accuracy: " ,accuracy , "%")





```
